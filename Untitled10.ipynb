```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_netflix = pd.read_csv("netflix_titles.csv")
```

```python
# Conferindo valores nulos e formato dos valores
print(df_netflix.isna())
df_netflix.info()
```

```python
# Verificando se tem valores duplicados
duplicados = df_netflix.duplicated().any()

if duplicados:
    print('Existem itens duplicados no DataFrame')
else:
    print('Não existem valores duplicados no DataFrame')
```

```python
# Transformando a coluna para string e depois para data
df_netflix['date_added'] = df_netflix['date_added'].astype(str)
df_netflix['date_added'] = pd.to_datetime(df_netflix['date_added'], format='%B %d, %Y', errors='coerce')

df_netflix['release_year'] = df_netflix['release_year'].astype(str)
df_netflix['release_year'] = pd.to_datetime(df_netflix['release_year'])

# extraindo apenas o ano da coluna de datas
df_netflix['release_year'] = df_netflix['release_year'].dt.year
```

```python
df_netflix.head()
```

```python
# Anos com mais lançamentos de titulos - Perguntas que podem ser respondidas com o grafico
 # 1° Quais foram os anos com mais lançamentos na Netflix?
 # 2° Houve alguma mudança significativa no número de lançamentos ao longo dos anos?


# Contagem de lançamento por ano - Total catalogo
contagem_anos = df_netflix['release_year'].value_counts()

# Ordenando os anos de lançamentos e selecionando o top 10
top_10 = contagem_anos.sort_values(ascending=False).head(10)

# ------------------------------------------------------------ #

# Contagem de lançamento por ano - Filmes
filmes_ano = df_netflix[df_netflix['type'] == 'Movie']
contagem_filmes_ano = filmes_ano['release_year'].value_counts()

# Ordenando os anos de lançamentos e selecionando o top 10
top_10_filmes = contagem_filmes_ano.sort_values(ascending=False).head(10)

# ------------------------------------------------------------ #

# Contagem de lançamento por ano - Series
series_ano = df_netflix[df_netflix['type'] == 'TV Show']
contagem_series_ano = series_ano['release_year'].value_counts()

# Ordenando os anos de lançamentos e selecionando o top 10
top_10_series = contagem_series_ano.sort_values(ascending=False).head(10)

# ------------------------------------------------------------ #

# Criando a figura e os subplots lado a lado
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plotando o primeiro gráfico (Total Catálogo)
axs[0].bar(top_10.index, top_10.values)
axs[0].set_title('Total Catálogo')

# Plotando o segundo gráfico (Filmes)
axs[1].bar(top_10_filmes.index, top_10_filmes.values, color = 'orange')
axs[1].set_title('Filmes')

# Plotando o terceiro gráfico (Séries)
axs[2].bar(top_10_series.index, top_10_series.values, color = 'green')
axs[2].set_title('Séries')

# Ajustando o layout para evitar sobreposição
plt.tight_layout()

# Exibindo os gráficos
plt.show()
```

```python
# Contagem de filmes por paises
top_5_paises = df_netflix['country'].value_counts().head(5)

# Criando grafico
plt.figure(figsize=(6,6)) # Tamanho do grafico
plt.pie(top_5_paises, labels = top_5_paises.index, autopct='%1.1f%%', startangle=140)
plt.title('Filmes por Países')

plt.show()
```

```python
## Compração de filmes e series - Respondendo as segunites perguntas
  #  1° Qual é a proporção de filmes e séries no catálogo da Netflix?
  #  2° Qual tipo de conteúdo tem maior representativeidade na plataforma?

# Contando quantidade de filmes e series
filmes = (df_netflix['type'] == 'Movie').sum()
series = (df_netflix['type'] == 'TV Show').sum()

# Contando o total de titulos no catalogo
total = len(df_netflix)

# Definindo as cores
cores = ['blue', 'orange', 'green']

# Criando o grafico de barras
barras = plt.bar(['Total', 'Filmes', 'Series'],[total, filmes, series], color=cores)

# Adicionando rótulos
plt.xlabel('Tipo')
plt.ylabel('Quantidade')
plt.title('Quantidade de Filmes e Series no Catalogo')

# Adicionando os valores no topo das barras
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2, altura, str(int(altura)), ha='center', va='bottom')

    # Adicionando o valor no topo da barra
    plt.text(barra.get_x() + barra.get_width() / 2, altura, str(int(altura)), ha='center', va='bottom')

plt.show()
```
